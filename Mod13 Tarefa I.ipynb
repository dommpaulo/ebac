2. Remova a variável menos significante e analise:

```python
# 1. Model adjustment, first part
y, x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + C(educacao) + C(estado_civil) + C(tipo_residencia) + idade + tempo_emprego + qt_pessoas_residencia ', data = df)
x
```

```python
sm.OLS(y, x).fit().summary()
```

#### Checking the most frequent category :

sexo -> F, tipo_renda -> Assalariado, educação -> secundário, estado_civil -> casado, tipo_residencia -> casa

```python
df.groupby(['tipo_residencia']).count()
```

```python
y, x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + C(educacao, Treatment(2)) + C(estado_civil) + C(tipo_residencia, Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia ', data = df)
```

```python
sm.OLS(y, x).fit().summary()
```

```python
# 2. Analysis of variables less meaningful
y, x = patsy.dmatrices('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego', data = df)
```

```python
sm.OLS(y, x).fit().summary()
```

#### Final model:
