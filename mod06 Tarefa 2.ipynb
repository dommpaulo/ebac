####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc = pd.read_csv('sinasc_2019.csv')
print(sinasc.shape)
```

```python
plt.close('all')
sinasc['PESO'].plot.hist(bins = 25)
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all')
sns.boxplot(y="PESO", data=sinasc)
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all')
sns.displot(data = sinasc
            ,alpha = .25
            ,x = 'IDADEMAE'
            ,bins = 25
            ,element = 'step'
            ,kde = True
           )
```
