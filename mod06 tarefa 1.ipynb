#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sinasc['APGAR5_MED'] = round((sinasc['APGAR5'] / sinasc['CONSULTAS']), 2)
sinasc_apgar5 = sinasc[['APGAR5' , 'CONSULTAS', 'APGAR5_MED']].copy()
sinasc_apgar5.head()
```

```python
plt.close('all')
sns.barplot(x='CONSULTAS', y='APGAR5_MED', data=sinasc_apgar5)
```
